```python
import polars as pl
import polars.selectors as cs

import altair as alt
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_path = '/Users/zygimantas/Documents/DataSets/Dataset salary 2024.csv'
```

```python
df = pl.read_csv(df_path)
```

```python
df
```

```python
df.group_by(
    pl.col('job_title')
).agg(
    pl.sum('salary').alias('total_salary'),
    pl.mean('salary').alias('avg_salary'),
    pl.len().alias('count'),
).sort(
    'avg_salary', descending=False
).filter(
    pl.col('count') > 10
)
```

```python
df.select(
    pl.col('salary_in_usd').mean().round(2).alias('avg_salary'),
    pl.col('salary_in_usd').std().round(2).alias('std_salary'),
    pl.col('salary_in_usd').median().round(2).alias('median_salary'),
    pl.col('salary_in_usd').quantile(0.1).round(2).alias('q1'),
    pl.col('salary_in_usd').quantile(0.25).round(2).alias('q25'),
    pl.col('salary_in_usd').quantile(0.5).round(2).alias('q50'),
    pl.col('salary_in_usd').quantile(0.75).round(2).alias('q75'),
)
```
